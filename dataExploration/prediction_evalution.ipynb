```python
y1 = []
y2 = []
y3 = []
y4 = []
parameters = []
with open("parameters_temp_1stack.json") as jsondata:
    data = json.load(jsondata)
    for d in data:
        y1.append(data[d]['precision'])
        y2.append(data[d]['recall'])
        y3.append(data[d]['fbeta_score'])
        y4.append(data[d]['accuracy'])
        
color = sns.color_palette()
```

```python
hid1 = []
hid2 = []
batch1 = []
batch2 = []
with open("parameters_temp_1stack.json") as jsondata:
    data = json.load(jsondata)
    for d in data:
        if data[d]['hid'] == 128:
            hid1.append(data[d]['fbeta_score'])
        else:
            hid2.append(data[d]['fbeta_score'])
#         y1.append(data[d]['precision'])
#         y2.append(data[d]['recall'])
        if data[d]['batch'] == 256:
            batch1.append(data[d]['fbeta_score'])
        else:
            batch2.append(data[d]['fbeta_score'])
#         y4.append(data[d]['accuracy'])
```

```python
x = list(range(len(y1)))
# print(x)
plt.figure(figsize=(20,12))
sns.pointplot(x, y1, alpha=0.1, color=color[1])
sns.pointplot(x, y2, alpha=0.2, color=color[2])
sns.pointplot(x, y3, alpha=0.3, color=color[3])
# sns.pointplot(x, y4, alpha=0.4, color=color[6])

color_patch1 = mpatches.Patch(color=color[1], label="precision")
color_patch2 = mpatches.Patch(color=color[2], label="recall")
color_patch3 = mpatches.Patch(color=color[3], label="fbeta_score")
plt.legend(handles=[color_patch1,color_patch2,color_patch3])
plt.ylabel('Evaluation', fontsize=12)
plt.xlabel('Parameters combination', fontsize=12)
plt.title("Single Stack LSTM", fontsize=15)
plt.xticks(rotation='vertical')
plt.show()
# plt.savefig("LSTM1.png")
```

```python
drop1 = []
drop2 = []
drop3 = []
drop4 = []
with open("parameters_temp_1stack.json") as jsondata:
    data = json.load(jsondata)
    for d in data:
        if data[d]['drop'] == 0.0:
            drop1.append(data[d]['fbeta_score'])
        elif data[d]['drop'] == 0.1:
            drop2.append(data[d]['fbeta_score'])
        elif data[d]['drop'] == 0.25:
            drop3.append(data[d]['fbeta_score'])
        elif data[d]['drop'] == 0.5:
            drop4.append(data[d]['fbeta_score'])
```

```python
x1 = list(range(len(drop1)))
x2 = list(range(len(drop2)))
x3 = list(range(len(drop3)))
x4 = list(range(len(drop4)))

# print(x)
plt.figure(figsize=(20,12))

sns.pointplot(x1, drop1, alpha=0.1, color=color[1], label = "0.0")
sns.pointplot(x2, drop2, alpha=0.2, color=color[2],label = "0.1")
sns.pointplot(x3, drop3, alpha=0.1, color=color[3], label = "0.25")
sns.pointplot(x4, drop4, alpha=0.2, color=color[4], label = "0.5")

# sns.pointplot(x, y4, alpha=0.4, color=color[6])

plt.ylabel('Evaluation', fontsize=12)
plt.xlabel('Parameters combination', fontsize=12)
plt.title("Single Stack LSTM", fontsize=15)
plt.xticks(rotation='vertical')
plt.legend()
plt.show()
```
